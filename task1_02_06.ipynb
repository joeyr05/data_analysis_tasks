```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
df['Gender'].nunique()
```

```python
#mapping binary values
gender_map = {'M': 0, 'F': 1}
df['Gender'] = df['Gender'].map(gender_map)
noshow_map = {'No': 0, 'Yes': 1}
df['No-show'] = df['No-show'].map(noshow_map)
```

```python
df.head()
```

```python
# Convert both to datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# Getting the date columns
df['scheduled_date'] = df['ScheduledDay'].dt.date
df['appointment_date'] = df['AppointmentDay'].dt.date

# (Optional) Drop original combined columns if not needed
df.drop(['ScheduledDay', 'AppointmentDay'], axis=1, inplace=True)
```

```python
df['PatientId'] = df['PatientId'].astype('int64') #changing patient ID to int values
```

```python
#converting header to lower case
df.columns = df.columns.str.lower()
```
